```python
import pandas as pd
from dfhelper.viz import df_view, df_info_view, df_shape_view
from dfhelper.scout import summarize_df, summarize_dfs
```

```python
df1 = pd.DataFrame(
            {"A": [1, 0, 0, None],
             "B": [1, 1, 2, 2],
             "C": [None, None, None, None]}
        )

df2 = pd.DataFrame(
            {"A": [1, 5, 0, 10],
             "B": [1, 1, 2, 2],
             "C": [None, 4, 16, 101]}
        )
```

## Viz

```python
df_view(df1, df2, orientation='hor')
```

```python
titles = ('df1', 'df2')
df_info_view(df1, df2, titles=titles, color_line='red', title_alignment='center')
```

```python
df_shape_view(df1, df2, show_titles=False)
```

## Scout

```python
summarize_df(df1)
```

```python
_ = summarize_dfs(df1, df2)
```
